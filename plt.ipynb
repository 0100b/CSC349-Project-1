## Plotting Results.csv

```python
import pandas as pd
import matplotlib.pyplot as plt

data_path = "results.csv"
df = pd.read_csv(data_path)

# Size,Selection,Merge,Counting
plt.scatter(df["Size"], df["Selection"], label="Selection")
plt.scatter(df["Size"], df["Merge"], label="Merge")
plt.scatter(df["Size"], df["Counting"], label="Counting")
plt.title("Sorting Algo. Comparison")
plt.xlabel("Size")
plt.ylabel("Time")
plt.legend()
plt.show()
```
